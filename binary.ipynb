Decimal to binary

```python
def dectobin(n:int):
    if n==0:
        return "0"
    bin=""
    while n>0:
        digit=n%2
        bin=str(digit)+bin
        n=n//2
    return bin
print(dectobin(10))
```

Binary to Decimal

```python
def bintodec(n:int):
    if n == 0:
        return 0
    ans = 0
    n = str(n)
    length = len(n)
    for i in range(length):
        curr = int(n[i])
        ans += curr * (2 ** (length - i - 1))
    return ans

print(bintodec(10))
```

Decimal to Octal

```python
def dectooct(n):
    if n==0:
        return "0"
    oct=""
    while n>0:
        digit=n%8
        oct=str(digit)+oct
        n=n//8
    return oct
print(dectooct(100))
```

Octal to decimal

```python
def octtodec(n):
    if n==0:
        return 0
    n=str(n)
    length=len(n)
    ans=0
    for i in range(len(n)):
        ans+=int(n[i])*(8**(length-i-1))
    return ans
print(octtodec(144))
```

Decimal to hex

```python
def dectohex(n):
    if n==0:
        return "0"
    hex_val="0123456789ABCDEF"
    hex=""
    while n>0:
        digit=n%16
        hex=hex_val[digit]+hex
        n=n//16
    return hex
print(dectohex(1003424))
```

Hex to Decimal

```python
def hextodec(n):
    if n==0:
        return 0
    n=str(n)
    length=len(n)
    ans=0
    hex_map = "0123456789ABCDEF"
    for i in range(length):
        char=n[i]
        curr=hex_map.find(char)
        ans+=curr*(16**(length-i-1))
    return ans
print(hextodec('F100'))
```

Recursion conversion

```python
def oct_to_dec(oct_str,power=0):
    if not oct_str:
        return 0
    return int(oct_str[-1])*(8**power) + oct_to_dec(oct_str[:-1],power+1)

def dec_to_hex(n):
    hex_chars='0123456789ABCDEF'
    if n<16:
        return hex_chars[n]
    return dec_to_hex(n//16) + hex_chars[n%16]

def oct_to_hex(oct_str):
    d=oct_to_dec(oct_str)
    return dec_to_hex(d)

print(oct_to_hex('157'))
```

Binary wrapping

```python
def binary_wrap(num):
    binary = bin(num)[2:]
    while len(binary) > 4:
        msb = binary[:-4] 
        lsb = binary[-4:]
        
        # Add MSB and LSB (both binary → integer → add → binary again)
        wrapped = int(lsb, 2) + int(msb, 2)
        
        # Convert back to binary string
        binary = bin(wrapped)[2:]
    
    # Return the 4-bit result
    return binary.zfill(4)   # Ensures 4-bit format with leading zeros if needed


# Example:
print(binary_wrap(57))
```
