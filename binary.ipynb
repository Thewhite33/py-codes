Decimal to binary

```python
def dectobin(n:int):
    if n==0:
        return "0"
    bin=""
    while n>0:
        digit=n%2
        bin=str(digit)+bin
        n=n//2
    return bin
print(dectobin(13667))
```

Binary to Decimal

```python
def bintodec(n:int):
    if n == 0:
        return 0
    ans = 0
    n = str(n)
    length = len(n)
    for i in range(length):
        curr = int(n[i])
        ans += curr * (2 ** (length - i - 1))
    return ans

print(bintodec(3563))
```

Hexadecimal to decimal

```python
def hex_to_dec(hex_num):
    hex_num = str(hex_num).upper()
    dec = 0
    length = len(hex_num)
    hex_digits = "0123456789ABCDEF"
    for i in range(length):
        dec += hex_digits.index(hex_num[i]) * (16 ** (length - 1 - i))
    return dec

print(hex_to_dec("123ef"))
```

Binary wrapping

```python
def binary_wrap(num):
    # Convert number to binary string (remove '0b')
    binary = bin(num)[2:]
    
    # Keep wrapping until it fits in 4 bits
    while len(binary) > 4:
        msb = binary[:-4]    # All bits before last 4 bits
        lsb = binary[-4:]    # Last 4 bits
        
        # Add MSB and LSB (both binary → integer → add → binary again)
        wrapped = int(lsb, 2) + int(msb, 2)
        
        # Convert back to binary string
        binary = bin(wrapped)[2:]
    
    # Return the 4-bit result
    return binary.zfill(4)   # Ensures 4-bit format with leading zeros if needed


# Example:
print(binary_wrap(1000))
```
