Reverse Dictonary key orders

```python
def revdict(d:dict):
    keys=[]
    values=[]
    new_dict={}
    for k,v in d.items():
        keys.append(k)
        values.append(v)
    keys=keys[::-1]
    for key,value in zip(keys,values):
        new_dict[key]=value
    return new_dict

print(revdict({'we':4,'are':2,'best':5}))
```

Sort a dictionary
i/p -> {3:"three",1:"one",2:"two"}
o/p -> {1:"one",2:"two",3:"three"}

```python
def sortDict(dic:dict):
    keys=sorted(list(dic.keys()))
    newdict={}
    for key in keys:
        newdict[key]=dic[key]
    return newdict

print(sortDict({3:"three",1:"one",2:"two"}))
```

Generate anagrams in list

```python
import random as r
def generateAnagrams(lst:list):
    dic={}
    for words in lst:
        chars=list(words)
        r.shuffle(chars)
        new_word=''.join(chars)
        dic[words]=new_word
    return dic

print(generateAnagrams(["hello","hi","how","are"]))
```

Count frequencies of elements in a list

```python
def countFrequency(lst:list):
    freq={}
    for item in lst:
        freq[item]=freq.get(item,0)+1
    return freq
print(countFrequency(['apple', 'banana', 'apple', 'orange', 'banana']))
```

Reverse key value in dictionary

```python
def revdic(dic:dict):
    key=[]
    val=[]
    newdic={}
    for k,v in dic.items():
        key.append(k)
        val.append(v)
    for k,v in zip(key,val):
        newdic[v]=k
    return newdic

print(revdic({'a': 1, 'b': 2, 'c': 3}))
```

Dictionary intersection

```python
def dicintersect(dic1:dict,dic2:dict):
    result=dic1.copy()
    for k,v in dic2.items():
        result[k]=result.get(k,0)+v
    return result
print(dicintersect({'a': 10, 'b': 5}, {'b': 15, 'c': 20}))
```

Group words by length

```python
def group_words_by_length(words):
    result={}
    for word in words:
        length=len(word)
        if length in result:
            result[length].append(word)
        else:
            result[length]=[word]
    return result

words = ["hi", "to", "sun", "cat", "moon", "star"]
output = group_words_by_length(words)
print(output)
```

Most common word in a string

```python
def most_common_word(s:str):
    s=s.split(' ')
    freq={}
    for i in s:
        freq[i]=freq.get(i,0)+1
    return max(freq,key=freq.get)

print(most_common_word("apple orange apple mango orange apple"))
```

```python
def findavgdict(d:dict):
    nd={}
    for k,v in d.items():
        s=0
        for n in v:
            s+=n
        nd[k]=s/len(v)
    return nd

print(findavgdict({'Student1':[14,16,18],'Student2':[12,15,17],'Student3':[16,16,13]}))
```
