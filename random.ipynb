Flatten a list

```python
def flatten(lst):
    res=[]
    for i in lst:
        if isinstance(i, list):
            res.extend(flatten(i))
        else:
            res.append(i)
    return res

print(flatten([1,2,[23,45,[6,8]]]))
```

Longest Substring without repeating characters

```python
def longestsubstr(string):
    maxlen = 0
    n = len(string)
    
    for i in range(n):
        for j in range(i + 1, n + 1):
            substring = string[i:j]
            if len(set(substring)) == len(substring):
                maxlen = max(maxlen, len(substring))
    return maxlen


print(longestsubstr("abcabcd"))
```

Valid Palindrom

```python
def validParenthesis(s:str):
    st=[]
    for i in s:
        if(i == '('):
            st.append(')')
        elif(i == '['):
            st.append(']')
        elif(i == '{'):
            st.append('}')
        else:
            if not st or st.pop() != i:
                return False
    return len(st) == 0

print(validParenthesis('[(])'))
```

Find Pairs with a Target Sum

```python
def find_target(lst,target):
    seen=set()
    output=set()

    for num in lst:
        complement=target-num
        if complement in seen:
            pair=tuple(sorted((num,complement)))
            output.add(pair)
        seen.add(num)
    return output

print(find_target([8, 3, 2, 5, 7, 1],10))
```

Invert a Dictionary with List Values

```python
def invert_dict(d:dict[str,list[str]]) -> dict[str,list[str]]:
    inverted_dict={}
    for key, values in d.items():
        for item in values:
            if item not in inverted_dict:
                inverted_dict[item]=[]
                inverted_dict[item].append(key)
    return inverted_dict

my_dict = {
    'group1': ['a', 'b'],
    'group2': ['b', 'c'],
    'group3': ['a', 'd']
}
print(f"Original Dictionary: {my_dict}")
print(f"Inverted Dictionary: {invert_dict(my_dict)}")
```

Regular expression Matching

```python
def regular_exp(s:str,p:str) -> bool:
    if(len(p)==0):
        return len(s)==0
    first_char_matched=False
    if(len(s)>0 and p[0]==s[0] or p[0]=='.'):
        first_char_matched=True
    if(len(p)>1 and p[1]=='*'):
        not_take=regular_exp(s,p[2:])
        take=first_char_matched and regular_exp(s[1:],p)
        return not_take or take

    return first_char_matched and regular_exp(s[1:],p[1:])

print(regular_exp("aab","c*a*b"))
```

Implement Priority Queue

```python
class PriorityQueue:
    def __init__(self):
        self.queue=[]
    def insert(self,item,priority):
        self.queue.append((priority,item))
        self.queue.sort(reverse=True)
    def remove(self):
        if self.is_empty():
            return "Queue is Empty"
        return self.queue.pop(0)[1]
    def display(self):
        if self.is_empty():
            print("Queue is empty")
        else:
            for priority,item in self.queue:
                print("Item: ",{item},"Priority: ",{priority})
    def is_empty(self):
        return len(self.queue) == 0

pq=PriorityQueue()
while True:
    choice = input("\n1. Insert  2. Remove  3. Display  4. Exit\nEnter choice: ")
    if choice == "1":
        item = input("Enter Task/Item: ")
        priority = int(input("Enter Priority: "))
        pq.insert(item, priority)
        print("Inserted successfully!")

    elif choice == "2":
        print("Removed:", pq.remove())

    elif choice == "3":
        print("Current Priority Queue:")
        pq.display()

    elif choice == "4":
        print("Exiting...")
        break

    else:
        print("Invalid choice! Try again.")
```

Create pattern

```python
def pattern(s: str):
    index = 0
    step = 1
    result = []
    while index < len(s):
        part = s[index:index+step]
        part = "*" * (step - len(part)) + part
        result.append(part)
        index += step
        step += 1
    return "\n".join(result)

print(pattern("oranges"))
```

LinkedList Operation

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class LinkedList:
    def __init__(self):
        self.head=None
    def insert_at_end(self,data):
        new_node=Node(data)
        if not self.head:
            self.head=new_node
            return
        temp=self.head
        while temp.next:
            temp=temp.next
        temp.next=new_node
    def insert_at_beginnig(self,data):
        new_node=Node(data)
        new_node.next=self.head
        self.head=new_node
    def insert_at_any_pos(self,data,pos):
        if pos==0:
            self.insert_at_beginnig(data)
            return
        new_node=Node(data)
        temp=self.head
        for _ in range(pos-1):
            if not temp:
                return "Index not found"
            temp=temp.next
        new_node.next=temp.next
        temp.next=new_node
    def delete(self,key):
        temp=self.head
        if temp and temp.next==key:
            self.head=temp.next
            return
        prev=None
        while temp and temp.data!=key:
            prev=temp
            temp=temp.next
        if temp:
            prev.next=temp.next
    def delete_at(self,pos):
        if pos==0 and self.head:
            self.head=self.head.next
            return
        temp=self.head
        prev=None
        for _ in range(pos):
            prev = temp
            temp = temp.next
            if not temp:
                raise IndexError("Index out of range")
        prev.next = temp.next
    def search(self,key):
        temp=self.head
        pos=0
        while temp:
            if temp.data==key:
                return pos
            temp=temp.next
            pos+=1
        return -1
    
l1=LinkedList()
l1.insert_at_beginnig(10)
l1.insert_at_beginnig(20)
l1.insert_at_beginnig(30)
print(l1.search(5))
```

Queue Data Structure

```python
class Queue:
    def __init__(self,size):
        self.size=size
        self.front=-1
        self.rear=-1
        self.arr=[None]*size
    def enqueue(self,data):
        if self.rear == self.size-1:
            print("Queue overflow")
            return
        if self.front == -1:
            self.front=0
        self.rear+=1
        self.arr[self.rear]=data
    def dequeue(self):
        if self.front==self.size-1:
            return "Nothing to dequeue"
        if self.front == self.rear:  # only one element left
            self.front = self.rear = -1
        else:
            self.front+=1
    def display(self):
        if self.rear==-1:
            return "No elements in the queue"
        print(self.arr[self.front:self.rear+1])
q=Queue(5)
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)
q.dequeue()
# q.display()
```

Rotting Oranges

```python
def rotting_oranges(grid:list)->int:
    n=len(grid)
    m=len(grid[0])
    time=0
    directions=[[1,0],[-1,0],[0,1],[0,-1]]
    while True:
        to_rot=[]
        for i in range(n):
            for j in range(m):
                if grid[i][j]==2:
                    for v,u in directions:
                        ni=i+v
                        nj=j+u
                        if ni>=0 and ni<n and nj>=0 and nj<m and grid[ni][nj]==1:
                            to_rot.append((ni,nj))
        if not to_rot:
            break
        for i,j in to_rot:
            grid[i][j]=2
        time+=1
    for i in range(n):
        for j in range(m):
            if grid[i][j]==1:
                return -1
    return time

print(rotting_oranges([[2,1,1],[1,1,0],[0,1,1]]))
print(rotting_oranges([[2,1,1],[0,1,1],[1,0,1]]))
print(rotting_oranges([[0,2]]))
```

Compare Version Numbers

```python
def compareVersion(version1: str, version2: str):
    version1lst=version1.split('.')
    version2lst=version2.split('.')
    if len(version1lst)>len(version2lst):
        while len(version1lst)!=len(version2lst):
            version2lst.append('0')
    if len(version2lst)>len(version1lst):
        while len(version2lst)!=len(version1lst):
            version1lst.append('0') 
    i=0
    while i<len(version1lst):
        if int(version1lst[i])>int(version2lst[i]):
            return 1
        elif int(version1lst[i])<int(version2lst[i]):
            return -1
        i+=1
    return 0

print(compareVersion('1.01','1.001'))
```

Fraction to Recurring Decimal
Input: numerator = 1, denominator = 2
Output: "0.5"
Input: numerator = 4, denominator = 333
Output: "0.(012)"

```python
def fractionToDecimal(numerator: int, denominator: int):
    if(numerator==0):
        return "0"
    res=''
    integerDiv=int(numerator/denominator)
    res+=str(integerDiv)
    remain=int(numerator%denominator)
    if(remain==0):
        return res
    res+='.'
    return res

print(fractionToDecimal(1,2))
```

Longest Common Prefix

```python
def longestCommonPrefix(strs) -> str:
        if(len(strs)==0):
            return ""
        strs=sorted(strs)
        first=strs[0]
        last=strs[len(strs)-1]
        i=0
        ans=""
        while(i<len(first) and i<len(last)):
            if(first[i]==last[i]):
                ans+=first[i]
                i+=1
            else:
                break
        return ans

print(longestCommonPrefix(["flower","flow","flight"]))
```
