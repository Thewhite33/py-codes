Longest Palindromic String

```python
def isPalindrome(s:str):
    if(len(s)==1):
        return True
    rev=""
    for i in reversed(s):
        rev+=i
    if(s==rev):
        return True
    return False
def longestPalindromicStrng(s:str):
    longest=0
    longestPal=""
    for i in range(len(s)):
        for j in range(i,len(s)):
            tocheck=s[i:j+1]
            if(isPalindrome(tocheck)):
                longest=max(longest,len(tocheck))
                if(len(tocheck)==longest):
                    longestPal=tocheck
    return longestPal
longestPalindromicStrng("babad")
```

Longest substring without duplicate characters

```python
def longsubstr(s:str):
    maxlen=0
    for i in range(len(s)):
        seen=set()
        for j in range(i,len(s)):
            if s[j] in seen:
                break
            else:
                seen.add(s[j])
                maxlen=max(maxlen,j-i+1)


    return maxlen
print(longsubstr("aab"))
```

Roman to Decimal

```python
def romantodec(r):
    romans={'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}
    res=0
    for i in range(0,len(r)-1):
        if(romans.get(r[i])<romans.get(r[i+1])):
            res-=romans.get(r[i])
        else:
            res+=romans.get(r[i])
    res+=romans.get(r[-1])
    return res

print(romantodec('XXXII'))
```

```python
def intToRoman(num: int) -> str:
    roman = [(1000, "M"),
            (900, "CM"),
            (500, "D"),
            (400, "CD"),
            (100, "C"),
            (90, "XC"),
            (50, "L"),
            (40, "XL"),
            (10, "X"),
            (9, "IX"),
            (5, "V"),
            (4, "IV"),
            (1, "I")
        ]
        
    ans = ""
    for value, symbol in roman:
        while num >= value:
            ans += symbol
            num -= value
    return ans    

print(intToRoman(32))
```

```python
def reverseString(s: str):
    if len(s)<=1:
        return s
    return s[-1]+reverseString(s[:-1])

def reverse_each_word(sentence:str):
    if ' ' not in sentence:
        return reverseString(sentence)
    space_idx=sentence.find(' ')
    first=sentence[:space_idx]
    rest=sentence[space_idx+1:]
    return reverseString(first)+' '+reverse_each_word(rest)

print(reverse_each_word("Hello my name is rohit"))
```

Sort a sentence in decreasing order

```python
def sortSentence(s:str):
    lst=s.split(' ')
    newlst=[]
    while lst:
        smallest=lst[0]
        for word in lst:
            if len(word)<len(smallest):
                smallest=word
        newlst.append(smallest)
        lst.remove(smallest)
    dic={}
    for word in newlst:
        dic[word]=len(word)
    return dic

print(sortSentence("Hello my name is Rohit"))
```

Remove white spaces and lower characters

```python
def remove_and_display(s:str):
    lst=s.split()
    new_lst=[]
    removed_lst=[]
    for words in lst:
        if words.isupper():
            new_lst.append(words)
        elif words.islower():
            removed_lst.append(words)
    print("Original:", s)
    print("Removed:", removed_lst)

remove_and_display("Hello my name is Rohit")
```

First non-repeating characters

```python
def non_repeating(s):
    freq={}
    for ch in s:
        freq[ch]=freq.get(ch,0)+1
    for ch in s:
        if freq[ch]==1:
            return ch
    return None

print(non_repeating("swiss"))
```
