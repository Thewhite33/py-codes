```python
# Max of three
def maxofThree(x,y,z):
    max=x
    if(x<y):
        max=y
        if(y<z):
            max=z
    # elif(max<z):
    #     max=z
    return max
x=maxofThree(1,5,8)
print(x)
```

```python
#Sum of list
def sum(numbers):
    total=0
    for x in numbers:
        total+=x
    return total

print(sum((10,20,30)))
```

```python
# Reverse a string
def revstrs(string):
    return string[::-1]


print(revstrs('dakmi'))
```
