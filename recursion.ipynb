Insert at bottom of stack

```python
def insertAtBottom(s:list,num:int):
    if(len(s)==0):
        s.append(num)
        return
    top=s.pop()
    insertAtBottom(s,num)
    s.append(top)
    return s

print(insertAtBottom([3,2,1,5],100))
```

Reverse a stack using recursion

```python
def reverseStack(stack:list):
    if(len(stack)==0):
        return
    top=stack.pop()
    reverseStack(stack)
    insertAtBottom(stack,top)
    return stack

print(reverseStack([3,2,1,5]))
```

Sort a stack using recursion

```python
def insertSorted(stack:list,num:int):
    if(len(stack)==0 or stack[-1]<num):
        stack.append(num)
        return
    top=stack.pop()
    insertSorted(stack,num)
    stack.append(top)

def sortStack(stack:list):
    if(len(stack)==0):
        return
    top=stack.pop()
    sortStack(stack)
    insertSorted(stack,top)
    return stack

print(sortStack([5,-2,9,-7,3]))
```

Reverse a string using recursion

```python
def revString(s:str):
    if(len(s)<=1):
        return s
    return s[-1]+revString(s[:-1])
print(revString("rohit"))
```

Permutation of strings

```python
import random as rd 
def fac(x):
    res=1
    for i in range(1,x+1):
        res=res*i
    return res
def insert_in_list(s:str):
    l=[]
    for i in s:
        l.append(i)
    return l
def permutationString(s:str):
    l=insert_in_list(s)
    seen=set()
    while(len(seen)!=fac(len(s))):
        rd.shuffle(l)
        seen.add(''.join(l))
    return seen

print(permutationString('abc'))
```
