Write and Read operation in file

```python
file=open("ex.txt","w")
file.write("Hello my name is Rohit\nI am a good boy")
file.close()

fileread=open("ex.txt","r")
print(fileread.read())
fileread.close()
```

Appending in file

```python
file=open("ex.txt","a+")
file.write("\nNew line here\n")
```

Read from a file and count number of vowels and consonants and write result in another file

```python
r=open("ex.txt","r")
w=open("op.txt","w")
line=r.read()
vowel=consonant=0
vowels=set('aeiouAEIOU')
for i in line:
    if i in vowels:
        vowel+=1
    else:
        consonant+=1
w.writelines(f"Vowels-> {vowel}\n")
w.writelines(f"Consonants-> {consonant}")
```
