Write and Read operation in file

```python
file=open("ex.txt","w")
file.write("Hello my name is Rohit\nI am a good boy")
file.close()

fileread=open("ex.txt","r")
print(fileread.read())
fileread.close()
```

```python
with open("ex.txt","r") as f:
    for i in f:
        print(i,end="")
```

Appending in file

```python
file=open("ex.txt","a+")
file.write("\nNew line here\n")
```

Read from a file and count number of vowels and consonants and write result in another file

```python
r=open("ex.txt","r")
w=open("op.txt","w")
line=r.read()
vowel=consonant=0
vowels=set('aeiouAEIOU')
for i in line:
    if i in vowels:
        vowel+=1
    else:
        consonant+=1
w.writelines(f"Vowels-> {vowel}\n")
w.writelines(f"Consonants-> {consonant}")
```

```python
file_name="students.txt"

def add_student():
    name=input("Enter name")
    roll=input("Enter roll number")
    marks=input("Enter marks")

    with open(file_name,"w") as f:
        f.write(f"{name},{roll},{marks}\n")
    print("Successfully printed data")

add_student()
```
