Read from a file and count number of vowels and consonants and write result in another file

```python
r=open("ex.txt","r")
w=open("op.txt","w")
line=r.read()
vowel=consonant=0
vowels=set('aeiouAEIOU')
for i in line:
    if i in vowels:
        vowel+=1
    else:
        consonant+=1
w.writelines(f"Vowels-> {vowel}\n")
w.writelines(f"Consonants-> {consonant}")
```

```python
file_name="students.txt"

def add_student():
    name=input("Enter name")
    roll=input("Enter roll number")
    marks=input("Enter marks")

    with open(file_name,"a") as f:
        f.write(f"{name},{roll},{marks}\n")
    print("Successfully printed data")

def write_students(students):
    """Writes the full student list back to file."""
    with open(file_name, "w") as f:
        for s in students:
            f.write(",".join(s) + "\n")

def get_student():
    students=[]
    with open(file_name,"r") as f:
        for line in f:
            if line.strip():
                name,roll,marks=line.strip().split(",")
                students.append([name,roll,marks])
    return students

def display_students():
    students = get_student()
    if not students:
        print("No student records found.\n")
        return

    print("\n--- Student Database ---")
    print("Name\tRoll\tMarks")
    print("-----------------------------")

    for s in students:
        print(f"{s[0]}\t{s[1]}\t{s[2]}")
    print()

display_students()
```

Update marks and roll of student

```python
def add_student(name,roll,marks):
    with open("students.txt","a") as f:
        f.write(f"{name},{roll},{marks}\n")

def update_marks(roll,new_marks):
    students=[]
    with open("students.txt") as f:
        for line in f:
            data=line.strip().split(',')
            if data[1]==str(roll):
                data[2]=str(new_marks)
            students.append(",".join(data))
    
    with open("students.txt","w") as f:
        f.write("\n".join(students))

update_marks(348,90)
# add_student("Rohit",348,95)
add_student("Raman",323,80)
```

```python
import csv

input_file = "students.csv"
output_file = "report.txt"

students = []
skipped = 0

# Read CSV
with open(input_file, newline="") as f:
    reader = csv.DictReader(f)

    for row in reader:
        name = row.get("name", "").strip()
        marks = row.get("marks", "").strip()

        try:
            marks = float(marks)
            students.append((name, marks))
        except ValueError:
            skipped += 1

# Sort: marks DESC, then name ASC
students.sort(key=lambda x: (-x[1], x[0]))

# Compute class average
average = sum(m for _, m in students) / len(students)

# Write report
with open(output_file, "w") as f:
    f.write(f"Class average: {average:.2f}\n\n")

    f.write("Top 3:\n")
    for i, (name, marks) in enumerate(students[:3], start=1):
        f.write(f"{i}) {name} - {int(marks)}\n")

    f.write("\nAll students (sorted):\n")
    for name, marks in students:
        f.write(f"{name} - {int(marks)}\n")

    f.write(f"\nSkipped rows: {skipped}\n")
```
